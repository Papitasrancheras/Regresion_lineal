# Mexico DF: Host_acceptance_rate vs host_response_rate

## Room_type: Entire home/apt

```python
# Crear subconjuntos por el tipo de habitación  
mexico_rentire =mexico[mexico["room_type"] == "Entire home/apt"]
# Comprobamos el filtrado exitoso 
mexico_rentire["room_type"].value_counts()
```

### Regresión Lineal: “host_acceptance_rate vs host_response_rate”

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
y_var = mexico_rentire[["host_acceptance_rate"]]
x_var = mexico_rentire["host_response_rate"]
```

```python
# Creación del modelo:
# Convertimos x_var en una matriz de dos dimensiones
x_var = x_var.values.reshape(-1, 1)
#Se define model como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=x_var, y=y_var)
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

#### **Ecuación del modelo: y = 0.3969x + 0.612**

#### Evaluación del modelo

```python
#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
coef_Deter=model.score(X=x_var, y=y_var)
coef_Deter
```

```python
#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
coef_Correl=np.sqrt(coef_Deter)
coef_Correl
```

#### Predicciones 

```python
# Realizamos predicciones 
y_pred= model.predict(X=mexico_rentire[['host_response_rate']])
#Insertamos la columna de predicciones en el DataFrame
mexico_rentire.insert(0, 'Predicciones', y_pred)

#Visualizamos la gráfica comparativa entre el total real y el total predecido
sns.scatterplot(x='host_response_rate', y='host_acceptance_rate', color="blue", data=mexico_rentire)
sns.scatterplot(x='host_response_rate', y='Predicciones', color="red", data=mexico_rentire)
plt.ylim(0.5, 1)
plt.show()
```
